# IMPORTACIÓN DE LIBRERÍAS

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

# Configuración visual
plt.style.use("seaborn-v0_8")
pd.set_option("display.max_columns", None)
```

# CARGA DE DATOS

```python
X = pd.read_csv("training_features.csv")
y = pd.read_csv("training_target.csv")

# Unimos temporalmente para análisis exploratorio
df = pd.concat([X, y], axis=1)

print("Dimensiones del dataset:")
print(df.shape)

df.head()
```

```python
print("\nInformación general:")
df.info()

print("\nTipos de variables:")
print(df.dtypes.value_counts())
```

# IDENTIFICACIÖN DE VARIABLES

```python
num_cols = df.select_dtypes(include=["int64", "float64"]).columns
cat_cols = df.select_dtypes(include=["object"]).columns

print("Número de variables numéricas:", len(num_cols))
print("Número de variables categóricas:", len(cat_cols))

print("\nVariables numéricas:")
print(num_cols)

print("\nVariables categóricas:")
print(cat_cols)
```

## VALORES FALTANTES

```python
missing = df.isnull().sum()
missing = missing[missing > 0].sort_values(ascending=False)

print("Columnas con valores faltantes:")
print(missing)

print("\nPorcentaje de valores faltantes:")
print((missing / len(df)) * 100)
```

## ESTADÍSTICAS DESCRIPTIVAS

```python
df[num_cols].describe()
```

# DISTRIBUCIONES

```python
for col in num_cols:
    plt.figure(figsize=(6,4))
    sns.histplot(df[col], kde=True)
    plt.title(f"Distribución de {col}")
    plt.show()
```

# DETECCIÓN DE OUTLIERS

```python
for col in num_cols:
    plt.figure(figsize=(6,2))
    sns.boxplot(x=df[col])
    plt.title(f"Boxplot de {col}")
    plt.show()
```

# ANÁLISIS DEL TARGET

```python
target = y.columns[0]

print("Estadísticas del target:")
print(df[target].describe())

plt.figure(figsize=(6,4))
sns.histplot(df[target], kde=True)
plt.title(f"Distribución de {target}")
plt.show()
```

# MATRIZ DE CORRELACIÓN

```python
corr = df.corr(numeric_only=True)

plt.figure(figsize=(12,10))
sns.heatmap(corr, cmap="coolwarm", center=0)
plt.title("Matriz de correlación")
plt.show()
```

# CORRELACIÓN CON EL TARGET

```python
corr_target = corr[target].sort_values(ascending=False)

print("Correlación de variables con el target:")
print(corr_target)
```
